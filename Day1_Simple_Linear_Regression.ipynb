```python
# Step 1: Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 2: Create dataset
data = {
    "Size_sqft": [1000, 1500, 2000, 2500, 3000],
    "Price": [300000, 400000, 500000, 600000, 700000]
}
df = pd.DataFrame(data)

# Step 3: Define Features (X) and Target (y)
X = df[["Size_sqft"]]   # Double brackets â†’ keeps it as DataFrame
y = df["Price"]

# Step 4: Create model and fit
model = LinearRegression()
model.fit(X, y)

# Step 5: Prediction
predicted_price = model.predict([[2200]])  # Predict for 2200 sqft
print("Predicted Price:", predicted_price[0])

# Step 6: Plot data and regression line
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, model.predict(X), color="red", label="Regression Line")
plt.xlabel("Size (sqft)")
plt.ylabel("Price")
plt.legend()
plt.show()
```
